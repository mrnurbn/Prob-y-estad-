## 6. La siguiente celda contiene un ejemplo de cómo realizar un $boxplot$ ¿Qué representa este tipo de gráficas? Realiza una gráfica $boxplot$ con la subbase del punto anterior 

La gráfica boxplot también es conocida como diagrama cajas y bigotes, la cuál es considerada como una forma visual en la que se puede mostrar grupos de datos númericos a través de los cuartiles. En otras palabaras es un diagrama que nos muestra la distribución de los datos para una variable continua.
La manera en la que se interpreta está gráfica es la siguiente: la líne central de la caja es la mediana, los extremos nos indican los cuartiles (la longitud de la caja va ser la diferencia que existe entre dos cuartiles), la líneas que se extienden (también conocidas como bigotes) representan la varianza que se espera de los datos 

```python
import numpy as np
import pandas as pd 
```

```python
df = pd.DataFrame(data3, columns=['Energy','Popularity'])
```

```python
boxplot = df.boxplot(column=['Energy','Popularity'])
```
